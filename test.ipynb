```python
import tensorflow as tf

x = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
y = tf.constant([[2, 2, 2], [3, 3, 3]], dtype=tf.float32)
z = tf.constant(321, dtype=tf.float32)

@tf.custom_gradient
def custom_sin_plus_exp (x, y, z):
  def grad (upstream):
    return upstream * tf.cos(x), upstream * tf.exp(y), None
  return tf.sin(x) + tf.exp(y), grad

with tf.GradientTape() as tape:
  tape.watch(x)
  tape.watch(y)
  tape.watch(z)
  w = custom_sin_plus_exp(x, y, z)

grad = tape.gradient(w, (x, y, z))
print(w)
print(grad)

print('--------------------')

x = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
y = tf.constant([[2, 2, 2], [3, 3, 3]], dtype=tf.float32)

with tf.GradientTape() as tape:
  tape.watch(x)
  tape.watch(y)
  z = tf.sin(x) + tf.exp(y)

grad = tape.gradient(z, (x, y))
print(z)
print(grad)
```

```python
def rnnt_loss (logits, labels, time_lengths, label_lengths):
  pr = pr_loss(tf.nn.softmax(logits, axis=-1), labels, time_lengths, label_lengths)
  ret = tf.reduce_sum(pr)
  return ret

@tf.custom_gradient
def pr_loss (pr, labels, time_lengths, label_lengths):
  pr = pr
  batch_size = pr.shape[0]
  max_time_lengths = pr.shape[1]
  max_label_lengths = pr.shape[2]

  def get_alpha (pr, labels, time_lengths, label_lengths):
    alpha = np.zeros((batch_size, max_time_lengths, max_label_lengths), dtype=np.float32)
    for b in range(batch_size):
      alpha[b][0][0] = 1

      for t in tf.range(time_lengths[b]):
        for u in tf.range(label_lengths[b]):
          if t + 1 >= time_lengths[b] or u + 1 >= label_lengths[b]:
            continue
          alpha[b][t + 1][u] += alpha[b][t][u] * pr[b][t][u][0]
          alpha[b][t][u + 1] += alpha[b][t][u] * pr[b][t][u][labels[b][u + 1]]

    return alpha
  
  def get_beta (pr, labels, time_lengths, label_lengths):
    beta = np.zeros((batch_size, max_time_lengths, max_label_lengths), dtype=np.float32)
    for b in range(batch_size):
      beta[b][-1][-1] = pr[b][-1][-1][0]
      for t in reversed(range(time_lengths[b])):
        for u in reversed(range(label_lengths[b])):
          if t - 1 < 0 or u - 1 < 0:
            continue
          beta[b][t - 1][u] += beta[b][t][u] * pr[b][t][u][0]
          beta[b][t][u - 1] += beta[b][t][u] * pr[b][t][u][labels[b][u - 1]]
    
    return beta

  alpha = get_alpha(pr, labels, time_lengths, label_lengths)
  beta = get_beta(pr, labels, time_lengths, label_lengths)
  
  total_pr = alpha[:, 0, 0] * beta[:, 0, 0]
  print('total_pr', total_pr.shape)

  vocab_size = pr.shape[-1]
  def grad (upstream):
    ret = np.zeros((batch_size, max_time_lengths, max_label_lengths, vocab_size), dtype=np.float32)
    for b in range(batch_size):
      for t in range(time_lengths[b]):
        for u in range(label_lengths[b]):
          if u + 1 < label_lengths[b]:
            ret[b][t][u][labels[b][u + 1]] = upstream[b] * -alpha[b][t][u] / (total_pr[b] + 1e-9) * beta[b][t][u + 1]
          if t + 1 < time_lengths[b]:
            ret[b][t][u][0] = upstream[b] * -alpha[b][t][u] / (total_pr[b] + 1e-9) * beta[b][t + 1][u]
    print('grad ret: ', tf.reduce_sum(tf.reduce_sum(ret)))
    return [ret] + [None] * 3

  return -tf.math.log(total_pr + 1e-9), grad
```

```python
import tensorflow as tf
from tensorflow.python.ops.gen_array_ops import matrix_diag_part_v3

LOG_0 = float('-inf')

truth_log_pr = tf.math.log_softmax(tf.random.uniform((10, 4, 2), dtype=tf.float32), axis=-1)

time_size = tf.shape(truth_log_pr)[-2]
label_size = tf.shape(truth_log_pr)[-1]

padded_truth_log_pr = tf.reverse(truth_log_pr, axis=[-2])
padded_truth_log_pr = tf.pad(
  padded_truth_log_pr,
  [[0, 0], [0, 0], [tf.shape(padded_truth_log_pr)[-2] - 1, 0]],
  constant_values=LOG_0
)
truth_diag = tf.linalg.diag_part(
  padded_truth_log_pr,
  k=(0, tf.shape(padded_truth_log_pr)[-1] - 1),
  padding_value=LOG_0,
  align='LEFT_RIGHT'
)
truth_diag = tf.pad(truth_diag, [[0, 0], [0, 0], [0, 1]], constant_values=LOG_0)

blank_log_pr = tf.math.log_softmax(tf.random.uniform((10, 4, 2), dtype=tf.float32), axis=-1)

reversed_blank_log_pr = tf.reverse(blank_log_pr, axis=[-2])
padded_blank_log_pr = tf.pad(
  reversed_blank_log_pr,
  [[0, 0], [0, 0], [tf.shape(reversed_blank_log_pr)[-2] - 1, 0]],
  constant_values=LOG_0
)
blank_diag = tf.linalg.diag_part(
  padded_blank_log_pr,
  k=(0, tf.shape(padded_blank_log_pr)[-1] - 1),
  padding_value=LOG_0,
  align='LEFT_RIGHT'
)
blank_diag = tf.pad(blank_diag, [[0, 0], [0, 0], [1, 0]], constant_values=LOG_0)

initial_diag = tf.concat([
  -0.5 * tf.ones((tf.shape(blank_diag)[0], 1), dtype=tf.float32),
  LOG_0 * tf.ones((tf.shape(blank_diag)[0], tf.shape(blank_diag)[-1] - 2), tf.float32),
], axis=-1)

def step (a, x):
  t, b = x
  return (
    tf.reduce_logsumexp(
      tf.stack([
        tf.concat([a, LOG_0 * tf.ones((tf.shape(a)[0], 1), dtype=tf.float32)], axis=-1) + t,
        tf.concat([LOG_0 * tf.ones((tf.shape(a)[0], 1), dtype=tf.float32), a], axis=-1) + b
      ]), axis=0
    )[:, : -1]
  )

beta_diag = tf.concat([
  tf.expand_dims(initial_diag, axis=0),
  tf.scan(step,
    (tf.transpose(truth_diag, perm=[1, 0, 2]),
    tf.transpose(blank_diag, perm=[1, 0, 2])),
    initial_diag
  )
], axis=0)

beta_diag = tf.transpose(beta_diag, [1, 0, 2])

beta = tf.linalg.diag_part(beta_diag, k=(-label_size + 1, 0))
beta = tf.transpose(beta, [0, 2, 1])
beta = tf.reverse(beta, axis=[1, 2])

print('truth_log_pr: ', truth_log_pr[0])
print()

print('blank_log_pr: ', blank_log_pr[0])
print()

print('beta: ', beta[0])
print()
```

```python
import tensorflow as tf
from tensorflow.python.ops.gen_array_ops import matrix_diag_part_v3

LOG_0 = float('-inf')

truth_log_pr = tf.math.log_softmax(tf.random.uniform((10, 4, 2), dtype=tf.float32), axis=-1)

time_size = tf.shape(truth_log_pr)[-2]
label_size = tf.shape(truth_log_pr)[-1]

print('truth_log_pr: ', truth_log_pr)
print()

reversed_truth_log_pr = tf.reverse(truth_log_pr, axis=[-2])
padded_truth_log_pr = tf.pad(
  reversed_truth_log_pr,
  [[0, 0], [0, 0], [tf.shape(reversed_truth_log_pr)[-2] - 1, 0]],
  constant_values=LOG_0
)

print('padded_truth_log_pr: ', padded_truth_log_pr)
print()

truth_diag = tf.linalg.diag_part(
  padded_truth_log_pr,
  k=(0, tf.shape(padded_truth_log_pr)[-1] - 1),
  padding_value=LOG_0,
  align='LEFT_RIGHT'
)
truth_diag = tf.reverse(truth_diag, axis=[-2])

print('truth_diag: ', truth_diag)
print()

truth_diag = tf.pad(truth_diag, [[0, 0], [0, 0], [1, 0]], constant_values=LOG_0)

print('truth_diag: ', truth_diag)
print()

blank_log_pr = tf.math.log_softmax(tf.random.uniform((10, 4, 2), dtype=tf.float32), axis=-1)

print('blank_log_pr: ', blank_log_pr)
print()

reversed_blank_log_pr = tf.reverse(blank_log_pr, axis=[-2])
padded_blank_log_pr = tf.pad(
  reversed_blank_log_pr,
  [[0, 0], [0, 0], [tf.shape(reversed_blank_log_pr)[-2] - 1, 0]],
  constant_values=LOG_0
)

print('padded_blank_log_pr: ', padded_blank_log_pr)
print()

blank_diag = tf.linalg.diag_part(
  padded_blank_log_pr,
  k=(0, tf.shape(padded_blank_log_pr)[-1] - 1),
  padding_value=LOG_0,
  align='LEFT_RIGHT'
)
blank_diag = tf.reverse(blank_diag, axis=[-2])

print('blank_diag: ', blank_diag)
print()

blank_diag = tf.pad(blank_diag, [[0, 0], [0, 0], [0, 1]], constant_values=LOG_0)

print('blank_diag: ', blank_diag)
print()

print('truth, blank diags: ')
print(truth_diag)
print(blank_diag)
print()

initial_diag = tf.concat([
  LOG_0 * tf.ones((tf.shape(blank_diag)[0], tf.shape(blank_diag)[-1] - 2), tf.float32),
  tf.zeros((tf.shape(blank_diag)[0], 1), dtype=tf.float32)
], axis=-1)

print('initial_diag: ', initial_diag)

def step (a, x):
  t, b = x
  return (
    tf.reduce_logsumexp(
      tf.stack([
        tf.concat([LOG_0 * tf.ones((tf.shape(a)[0], 1), dtype=tf.float32), a], axis=-1) + t,
        tf.concat([a, LOG_0 * tf.ones((tf.shape(a)[0], 1), dtype=tf.float32)], axis=-1) + b
      ]), axis=0
    )[:, 1: ]
  )

alpha_diag = tf.concat([
  tf.expand_dims(initial_diag, axis=0),
  tf.scan(step, (tf.transpose(truth_diag, perm=[1, 0, 2]), tf.transpose(blank_diag, perm=[1, 0, 2])), initial_diag)
], axis=0)

alpha_diag = tf.transpose(alpha_diag, perm=[1, 0, 2])

print('alpha_diag: ', alpha_diag.shape)
print()

alpha_diag = tf.reverse(alpha_diag, axis=[-1])
alpha_diag = tf.transpose(alpha_diag, perm=[0, 2, 1])
alpha = tf.linalg.diag_part(alpha_diag, k=(0, label_size - 1))
alpha = tf.reverse(alpha, axis=[-2])
alpha = tf.transpose(alpha, perm=[0, 2, 1])

print('truth_log_pr: ', truth_log_pr[0])
print()

print('blank_log_pr: ', blank_log_pr[0])
print()

print('alpha: ', alpha[0])
print()
```
